```python
#palindrome

def palindrome(s):
    m=s[::-1]
    if m==s:
        print("palindrome")
    else:
        print("not palindrome")
n=input()
palindrome(n)
```

```python
#factorial

def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
       
n=int(input())
factorial(n)
        
```

```python
#fibonacci 

def fibonacci(n):
    a=0
    b=1
    print(a,b,end=" ")
    while(n-2):
        c=a+b
        print(c,end=" ")
        n=n-1
        a=b
        b=c
n=int(input())
fibonacci(n)
    
```

```python
#Armstrong number

def armstrong(n):
    sum=0
    temp=n
    while(temp>0):
        d=temp%10
        sum+=d**3
        temp//=10
    
    if n==sum:
        print("armstrong number")
    else:
        print("not an armstrong number")
n=int(input())
armstrong(n)
```

```python
#odd or even

def oe(n):
    if n%2==0:
        print("even")
    else:
        print("odd")
n=int(input())
oe(n)
```

```python
#leap year

def leap(n):
    if n%4==0:
        if n%100==0:
            if n%400==0:
                print("leap year")
            else:
                print("not a leap year")
        else:
            print("leap year")
    else:
        print("not a leap year")
n=int(input())
leap(n)
                
```

```python
#prime number

def prime(n):
    if n>1:
        for i in range(2,n):
            if n%i==0:
                print("not prime")
                break
        else:
            print("prime")
    else:
        print("not prime")
n=int(input())
prime(n)
        
```

```python
#Area of shapes

import math
pi=math.pi
def circle(r):
    return pi*r**2
def cube(s):
    return 6*s**2
def cylinder(r,h):
    return 2*pi*r*h + 2*pi*r**2
print(circle(5))
print(cube(3))
print(cylinder(2,5))

    
```
